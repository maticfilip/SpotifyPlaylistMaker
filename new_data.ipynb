```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA
```

```python
df=pd.read_csv("1msongs.csv")
df.head()
```

```python
df.columns
```

```python
drop_cols=["Unnamed:0","uri","track_href","analysis_url","type","fduration_ms"]
df=df.drop(columns=drop_cols, errors="ignore")
```

```python
df=df.drop_duplicates(subset="track_uri")
```

```python
print(df.isnull().sum())
```

```python
df.fillna(0)
```

```python
df.describe()
```

```python
features=["danceability","energy","key","loudness","speechiness","mode","acousticness","instrumentalness","liveness","valence","tempo"]

pca=PCA(n_components=3)
pca_features=pca.fit_transform(df[features])
df["pca_1"]=pca_features[:, 0]
df["pca_2"]=pca_features[:, 1]
df["pca_3"]=pca_features[:, 2]

plt.figure(figsize=(10,6))
plt.scatter(df["pca_1"], df["pca_2"], alpha=0.5)
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans=KMeans(n_clusters=5, random_state=42)
df["cluster"]=kmeans.fit_predict(df[features])

plt.figure(figsize=(10, 6))
plt.scatter(df['pca_1'], df['pca_2'], c=df['cluster'], cmap='viridis')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('Song Clusters')
plt.colorbar(label='Cluster')
plt.show()
```
